# 03. Determining how to merge dataframes

The common column both dataframes share is product_id.
I'd like to test the merge first using an outer join

```python
pd.merge(df_ords_prods_comb, df_prods, on = 'product_id', how = 'outer')
```

```python
# Running another test merge, this time using a left join
pd.merge(df_ords_prods_comb, df_prods, on = 'product_id', how = 'left')
```

# 04. Merging dataframes

```python
# Merging dataframes using left join 
df_ords_prods_merged = df_ords_prods_comb.merge(
    df_prods, on = 'product_id', 
    how = 'left')
```

```python
# Confirming sucessful merge using merge flag
df_ords_prods_merged['_merge'].value_counts()
```
